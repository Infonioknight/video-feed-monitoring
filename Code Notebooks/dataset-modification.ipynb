## Guide for the below code
*  **'INPUT DIR'** is the path to the folder containing the **train**, **test** and **validation** folders.
*   Each of those folders contain an **images** and **labels** folder.
*   The below code modifies the text files in the **labels** folder according to the modifications mentioned in [this data-card](https://www.kaggle.com/datasets/gauravsekar/construction-site-safety-image-dataset-roboflow/data)

```python
import os

INPUT_DIR = 'path/to/your/input/folder'

for mode in ['train', 'valid', 'test']:

    path = os.path.join(INPUT_DIR, mode, 'labels')

    for file in os.listdir(path):
        with open(os.path.join(path, file), 'r+') as f:
            lines = f.readlines()

            f.seek(0)
            f.truncate(0)

            classes = ['Hardhat', 'NO-Hardhat', 'NO-Safety Vest', 'Person', 'Safety Vest', 'NOT-Person']

            new_lines = []
            for line in lines:
                if line[0] == '1' or line[0] == '3':
                    continue
                elif line[0] == '0':
                    new_lines.append(line)
                elif line[0] == '2':
                    new_lines.append('1' + line[1:])
                elif line[0] == '4':
                    new_lines.append('2' + line[1:])
                elif line[0] == '5':
                    new_lines.append('3' + line[1:])
                elif line[0] == '7':
                    new_lines.append('4' + line[1:])
                elif line[0] in ['6', '8', '9']:
                    new_lines.append('5' + line[1:])
            
            f.writelines(new_lines)
            f.close()
```
